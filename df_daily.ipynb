```python
def compute_baseline(df, col_date, series_id, cols_handler, n_past_values=5):
    for label_to_pred in cols_handler.labels_to_pred:
        steps = cols_handler.steps[label_to_pred]
        for step in steps:
            col_name = f"bl_{label_to_pred}_step{step}"
            cols_handler.add_col(label_to_pred, step, "baseline", col_name)

        baseline_cols = cols_handler.cols_type('baseline')[label_to_pred]
        df = df.sort_values(by=[series_id, col_date])

        val_baseline_max = df.groupby(series_id)[label_to_pred].apply(
            lambda x: x.rolling(n_past_values, 1).max().shift()).values
        val_baseline_min = df.groupby(series_id)[label_to_pred].apply(
            lambda x: x.rolling(n_past_values, 1).min().shift()).values
        val_baseline_sum = df.groupby(series_id)[label_to_pred].apply(
            lambda x: x.rolling(n_past_values, 1).sum().shift()).values

        val_baseline = (val_baseline_sum - val_baseline_min - val_baseline_max) / (n_past_values - 2)

        val_baseline = np.repeat(val_baseline[:, None], len(baseline_cols), axis=1)
        df[baseline_cols] = val_baseline
        df[baseline_cols] = df[baseline_cols].clip(0)

    df = df.dropna(subset=list(cols_handler.cols_type('baseline').values())[0])
    return df, cols_handler
```

```python
import re
import pandas as pd
import numpy as np
import scipy as sp
from scipy.spatial.distance import pdist
import sys
import warnings
import sklearn
import importlib
import copy
from contextlib import contextmanager
import sys, os


if (sys.version_info < (3, 0)):
    warnings.warn("As of version 0.29.0 shap only supports Python 3 (not 2)!")

import_errors = {}

def assert_import(package_name):
    global import_errors
    if package_name in import_errors:
        msg,e = import_errors[package_name]
        print(msg)
        raise e

def record_import_error(package_name, msg, e):
    global import_errors
    import_errors[package_name] = (msg, e)


def shapley_coefficients(n):
    out = np.zeros(n)
    for i in range(n):
        out[i] = 1 / (n * sp.special.comb(n-1,i))
    return out


def convert_name(ind, shap_values, input_names):
    if type(ind) == str:
        nzinds = np.where(np.array(input_names) == ind)[0]
        if len(nzinds) == 0:
            # we allow rank based indexing using the format "rank(int)"
            if ind.startswith("rank("):
                return np.argsort(-np.abs(shap_values).mean(0))[int(ind[5:-1])]

            # we allow the sum of all the SHAP values to be specified with "sum()"
            # assuming here that the calling method can deal with this case
            elif ind == "sum()": 
                return "sum()"
            else:
                raise ValueError("Could not find feature named: " + ind)
        else:
            return nzinds[0]
    else:
        return ind

def potential_interactions(shap_values_column, shap_values_matrix):
    """ Order other features by how much interaction they seem to have with the feature at the given index.

    This just bins the SHAP values for a feature along that feature's value. For true Shapley interaction
    index values for SHAP see the interaction_contribs option implemented in XGBoost.
    """

    # ignore inds that are identical to the column 
    ignore_inds = np.where((shap_values_matrix.values.T - shap_values_column.values).T.std(0) < 1e-8)
    
    values = shap_values_matrix.values
    X = shap_values_matrix.data

    if X.shape[0] > 10000:
        a = np.arange(X.shape[0])
        np.random.shuffle(a)
        inds = a[:10000]
    else:
        inds = np.arange(X.shape[0])

    x = shap_values_column.data[inds]
    srt = np.argsort(x)
    shap_ref = shap_values_column.values[inds]
    shap_ref = shap_ref[srt]
    inc = max(min(int(len(x) / 10.0), 50), 1)
    interactions = []
    for i in range(X.shape[1]):
        encoded_val_other = encode_array_if_needed(X[inds, i][srt], dtype=np.float)

        val_other = encoded_val_other
        v = 0.0
        if not (i in ignore_inds or np.sum(np.abs(val_other)) < 1e-8):
            for j in range(0, len(x), inc):
                if np.std(val_other[j:j + inc]) > 0 and np.std(shap_ref[j:j + inc]) > 0:
                    v += abs(np.corrcoef(shap_ref[j:j + inc], val_other[j:j + inc])[0, 1])
        val_v = v

        val_other = np.isnan(encoded_val_other)
        v = 0.0
        if not (i in ignore_inds or np.sum(np.abs(val_other)) < 1e-8):
            for j in range(0, len(x), inc):
                if np.std(val_other[j:j + inc]) > 0 and np.std(shap_ref[j:j + inc]) > 0:
                    v += abs(np.corrcoef(shap_ref[j:j + inc], val_other[j:j + inc])[0, 1])
        nan_v = v

        interactions.append(max(val_v, nan_v))

    return np.argsort(-np.abs(interactions))


def approximate_interactions(index, shap_values, X, feature_names=None):
    """ Order other features by how much interaction they seem to have with the feature at the given index.

    This just bins the SHAP values for a feature along that feature's value. For true Shapley interaction
    index values for SHAP see the interaction_contribs option implemented in XGBoost.
    """

    # convert from DataFrames if we got any
    if str(type(X)).endswith("'pandas.core.frame.DataFrame'>"):
        if feature_names is None:
            feature_names = X.columns
        X = X.values

    index = convert_name(index, shap_values, feature_names)

    if X.shape[0] > 10000:
        a = np.arange(X.shape[0])
        np.random.shuffle(a)
        inds = a[:10000]
    else:
        inds = np.arange(X.shape[0])

    x = X[inds, index]
    srt = np.argsort(x)
    shap_ref = shap_values[inds, index]
    shap_ref = shap_ref[srt]
    inc = max(min(int(len(x) / 10.0), 50), 1)
    interactions = []
    for i in range(X.shape[1]):
        encoded_val_other = encode_array_if_needed(X[inds, i][srt], dtype=np.float)

        val_other = encoded_val_other
        v = 0.0
        if not (i == index or np.sum(np.abs(val_other)) < 1e-8):
            for j in range(0, len(x), inc):
                if np.std(val_other[j:j + inc]) > 0 and np.std(shap_ref[j:j + inc]) > 0:
                    v += abs(np.corrcoef(shap_ref[j:j + inc], val_other[j:j + inc])[0, 1])
        val_v = v

        val_other = np.isnan(encoded_val_other)
        v = 0.0
        if not (i == index or np.sum(np.abs(val_other)) < 1e-8):
            for j in range(0, len(x), inc):
                if np.std(val_other[j:j + inc]) > 0 and np.std(shap_ref[j:j + inc]) > 0:
                    v += abs(np.corrcoef(shap_ref[j:j + inc], val_other[j:j + inc])[0, 1])
        nan_v = v

        interactions.append(max(val_v, nan_v))

    return np.argsort(-np.abs(interactions))

def encode_array_if_needed(arr, dtype=np.float64):
    try:
        return arr.astype(dtype)
    except ValueError:
        unique_values = np.unique(arr)
        encoding_dict = {string: index for index, string in enumerate(unique_values)}
        encoded_array = np.array([encoding_dict[string] for string in arr], dtype=dtype)
        return encoded_array

def sample(X, nsamples=100, random_state=0):
    if nsamples >= X.shape[0]:
        return X
    else:
        return sklearn.utils.resample(X, n_samples=nsamples, random_state=random_state)

def safe_isinstance(obj, class_path_str):
    """
    Acts as a safe version of isinstance without having to explicitly
    import packages which may not exist in the users environment.

    Checks if obj is an instance of type specified by class_path_str.

    Parameters
    ----------
    obj: Any
        Some object you want to test against
    class_path_str: str or list
        A string or list of strings specifying full class paths
        Example: `sklearn.ensemble.RandomForestRegressor`

    Returns
    --------
    bool: True if isinstance is true and the package exists, False otherwise
    """
    if isinstance(class_path_str, str):
        class_path_strs = [class_path_str]
    elif isinstance(class_path_str, list) or isinstance(class_path_str, tuple):
        class_path_strs = class_path_str
    else:
        class_path_strs = ['']
    
    # try each module path in order
    for class_path_str in class_path_strs:
        if "." not in class_path_str:
            raise ValueError("class_path_str must be a string or list of strings specifying a full \
                module path to a class. Eg, 'sklearn.ensemble.RandomForestRegressor'")

        # Splits on last occurence of "."
        module_name, class_name = class_path_str.rsplit(".", 1)

        # here we don't check further if the model is not imported, since we shouldn't have
        # an object of that types passed to us if the model the type is from has never been
        # imported. (and we don't want to import lots of new modules for no reason)
        if module_name not in sys.modules:
            continue

        module = sys.modules[module_name]
        
        #Get class
        _class = getattr(module, class_name, None)
        
        if _class is None:
            continue
        
        if isinstance(obj, _class):
            return True

    return False


def format_value(s, format_str):
    """ Strips trailing zeros and uses a unicode minus sign.
    """

    if not issubclass(type(s), str):
        s = format_str % s
    s = re.sub(r'\.?0+$', '', s)
    if s[0] == "-":
        s = u"\u2212" + s[1:]
    return s

# From: https://groups.google.com/forum/m/#!topic/openrefine/G7_PSdUeno0
def ordinal_str(n):
    """ Converts a number to and ordinal string. 
    """
    return str(n) + {1: 'st', 2: 'nd', 3: 'rd'}.get(4 if 10 <= n % 100 < 20 else n % 10, "th")

class OpChain():
    """ A way to represent a set of dot chained operations on an object without actually running them.
    """

    def __init__(self, root_name=""):
        self._ops = []
        self._root_name = root_name
    
    def apply(self, obj):
        """ Applies all our ops to the given object.
        """
        for o in self._ops:
            op,args,kwargs = o
            if args is not None:
                obj = getattr(obj, op)(*args, **kwargs)
            else:
                obj = getattr(obj, op)
        return obj

    def __call__(self, *args, **kwargs):
        """ Update the args for the previous operation.
        """
        new_self = OpChain(self._root_name)
        new_self._ops = copy.copy(self._ops)
        new_self._ops[-1][1] = args
        new_self._ops[-1][2] = kwargs
        return new_self
        
    def __getitem__(self, item):
        new_self = OpChain(self._root_name)
        new_self._ops = copy.copy(self._ops)
        new_self._ops.append(["__getitem__", [item], {}])
        return new_self

    def __getattr__(self, name):
        # Don't chain special attributes
        if name.startswith("__") and name.endswith("__"):
            return None
        new_self = OpChain(self._root_name)
        new_self._ops = copy.copy(self._ops)
        new_self._ops.append([name, None, None])
        return new_self

    def __repr__(self):
        out = self._root_name
        for o in self._ops:
            op,args,kwargs = o
            out += "."
            out += op
            if (args is not None and len(args) > 0) or (kwargs is not None and len(kwargs) > 0):
                out += "("
                if args is not None and len(args) > 0:
                    out += ", ".join([str(v) for v in args])
                if kwargs is not None and len(kwargs) > 0:
                    out += ", " + ", ".join([str(k)+"="+str(kwargs[k]) for k in kwargs.keys()])
                out += ")"
        return out

# https://thesmithfam.org/blog/2012/10/25/temporarily-suppress-console-output-in-python/
@contextmanager
def suppress_stderr():
    with open(os.devnull, "w") as devnull:
        old_stderr = sys.stderr
        sys.stderr = devnull
        try:
            yield
        finally:
            sys.stderr = old_stderr
```

```python
#HOW TO COMPUTE INTERACTION
def compute_interactions(shap_values_column, shap_values_matrix):
    """ 
    Order other features by how much interaction they seem to have with the feature at the given index.

    This just bins the SHAP values for a feature along that feature's value. For true Shapley interaction
    index values for SHAP see the interaction_contribs option implemented in XGBoost.
    """

    # ignore inds that are identical to the column 
    ignore_inds = np.where((shap_values_matrix.values.T - shap_values_column.values).T.std(0) < 1e-8)
    
    values = shap_values_matrix.values
    X = shap_values_matrix.data

    if X.shape[0] > 10000:
        a = np.arange(X.shape[0])
        np.random.shuffle(a)
        inds = a[:10000]
    else:
        inds = np.arange(X.shape[0])

    x = shap_values_column.data[inds]
    srt = np.argsort(x)
    shap_ref = shap_values_column.values[inds]
    shap_ref = shap_ref[srt]
    inc = max(min(int(len(x) / 10.0), 50), 1)
    interactions = []
    for i in range(X.shape[1]):
        encoded_val_other = encode_array_if_needed(X[inds, i][srt], dtype=np.float)

        val_other = encoded_val_other
        v = 0.0
        if not (i in ignore_inds or np.sum(np.abs(val_other)) < 1e-8):
            for j in range(0, len(x), inc):
                if np.std(val_other[j:j + inc]) > 0 and np.std(shap_ref[j:j + inc]) > 0:
                    v += abs(np.corrcoef(shap_ref[j:j + inc], val_other[j:j + inc])[0, 1])
        val_v = v

        val_other = np.isnan(encoded_val_other)
        v = 0.0
        if not (i in ignore_inds or np.sum(np.abs(val_other)) < 1e-8):
            for j in range(0, len(x), inc):
                if np.std(val_other[j:j + inc]) > 0 and np.std(shap_ref[j:j + inc]) > 0:
                    v += abs(np.corrcoef(shap_ref[j:j + inc], val_other[j:j + inc])[0, 1])
        nan_v = v

        interactions.append(max(val_v, nan_v))

    return interactions

    
# HOW TO COMPUTE INTERACTION MATRIX
def compute_interaction_matrix(shap_values,data):
  '''
  Given an ndarray(n_sample,n_features) of shap values and the the dataframe
  is returned a dataframe containing the pairwise interaction_matrix
  '''

  interaction_matrix =  []
  for el in data.columns:
    try:
      interaction_matrix.append(compute_interactions(shap_values[:, el], shap_values))
    except:
      interaction_matrix.append(np.zeros(len(data.columns)))
  interaction_matrix = pd.DataFrame(interaction_matrix,index = data.columns , columns = data.columns)


def plot_matrix(matrix,columns, title = 'MATRIX'):
  '''
  Given an input Dataframe encoding any kinf of interaction matrix (correlation,shap interactions,...)
  with shape(n_features,n_features), is plotted in a fancy way with matplotlib.pyplot
  '''
  fig, ax = plt.subplots(figsize =(8,8))
  im = ax.imshow(matrix)

  # Show all ticks and label them with the respective list entries
  ax.set_xticks(np.arange(len(columns)))
  ax.set_xticklabels(list(columns), fontsize = 13)

  ax.set_yticks(np.arange(len(columns)))
  ax.set_yticklabels(list(columns),fontsize = 13)

  # Rotate the tick labels and set their alignment.
  plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
          rotation_mode="anchor")

  # Loop over data dimensions and create text annotations.
  for i in range(len(columns)):
      for j in range(len(columns)):
          text = ax.text(j, i, round(matrix.to_numpy()[i, j],2),
                        ha="center", va="center", color="w", fontsize = 13)

  ax.set_title(title)
  fig.tight_layout()
  plt.show()
```

```python
def compute_adwin(data, date_col, col_to_check, product_ids, verbose = True):
  '''
  Given a univariate distribution of the variable (col_to_check) wrt one or multiples ids, 
  compute ADWIN index.
  A visualaization plot is computed if verbose = True (default).
  the list of warning indexes (index_adwin) is returned
  
  '''
  if type(product_ids) == int :
    product_ids = [product_ids]

  for j, el in enumerate(product_ids):
    adwin = ADWIN()
    indexes_adwin = []
    data_stream = data.loc[data['id'] == el,col_to_check]

    for i, v in enumerate(data_stream):
        in_drift_adwin, in_warning_adwin = adwin.update(v) #ADWIN
        if in_drift_adwin:
          indexes_adwin.append(i)
          adwin.reset()

    temp_adwin = data.iloc[indexes_adwin,:][date_col]
    if verbose:
      for el in indexes_adwin:
        print(f"Change detected at indexes {el}, input value: {data_stream.iloc[el]} \n")
      plt.figure(figsize = (12,5))
      plt.plot(data.loc[data['id'] == el, date_col], data_stream)
      for date in temp_adwin:
        plt.axvline(datetime.datetime(date.year,date.month, date.day), color = 'red', linestyle = '-')
  return indexes_adwin

def compute_kswin(data, date_col, col_to_check, product_ids, verbose = True):
  '''
  Given a univariate distribution of the variable (col_to_check) wrt one or multiples ids, 
  compute KSWIN index.
  A visualaization plot is computed if verbose = True (default).
  the list of warning indexes (index_kswin) is returned
  
  '''
  if type(product_ids) == int :
    product_ids = [product_ids]

  for j, el in enumerate(product_ids):
    kswin = KSWIN()
    indexes_kswin = []
    data_stream = data.loc[data['id'] == el,col_to_check]

    for i, v in enumerate(data_stream):
        in_drift_kswin, in_warning_kswin = kswin.update(v) #ADWIN
        if in_drift_kswin:
          indexes_kswin.append(i)
          kswin.reset()

    temp_kswin = data.iloc[indexes_kswin,:]['dat_trasporto']
    if verbose:
      for el in indexes_kswin:
        print(f"Change detected at indexes {el}, input value: {data_stream.iloc[el,:]} \n")

      plt.figure(figsize = (12,5))
      plt.plot(data.loc[data['id'] == el,'dat_trasporto'], data_stream)
      for date in temp_kswin:
        plt.axvline(datetime.datetime(date.year,date.month, date.day), color = 'red', linestyle = '-')
  return indexes_kswin
```

```python
def dataset_lda(data1,data2):
  '''
  Given two datasets data1, data2, them are concatenated and a column y is generated,
  where data[i,y] = 1 if is from data 1 and 0 ow
  '''
  labels = np.append(np.ones(len(data1)),np.zeros(len(data2)))
  if len(data1.shape) > 1:
    data = pd.concat([data1,data2])
    data['y'] = labels
  else:
    data = pd.concat([data1,data2])
    labels = np.append(np.ones(len(data1)),np.zeros(len(data2)))
    data = pd.DataFrame(np.array([data,labels]).T)
  return data.iloc[:,:-1], data.iloc[:,-1]

def lda_comp(data1,data2):
  '''
  Given two dataset is computed LDA(1).
  Is returned the fitted LDA and a dataframe containing two columns ['LDA_scores','labels'] for each sample '''

  lda = LDA()
  X,y = dataset_lda(data1,data2)
  try:
    X_transf = np.squeeze(lda.fit_transform(X,y),1)
  except: 
    X_transf = np.zeros(len(y))

  df = pd.DataFrame(np.array([X_transf,y]).T, columns = ['scores','labels'])
  return lda, df

def accuracy_test(accuracy,n, null_mean = 0.9, alpha = 0.05, verbose = True):
  '''
  Given an accuracy score, is tested if it's value in population is = to null_mean at a
  significance level of 0.05 - Morever is returned the C.I, of the computed accuracy in population
  If verbose = True a summary is printed.
  '''
  n_successes = int(n*accuracy)
  p_value = binom_test(n_successes,n, p = null_mean)
  confidence_interval = proportion_confint(int(n_successes),n,alpha = alpha)
  if verbose:
    print(f'''---- Accuracy test -----

test accuracy: {round(accuracy,2)}

The p_value of statistic test with Ho: p = {null_mean} with a significative value of {alpha} is: {p_value}
          
With a probability of {1- alpha} the accuracy in population is contained the C.I {confidence_interval}

''')

  return p_value, confidence_interval

def distribution_test_lda_1(data1,data2,null_mean = 0.7, n_neigh = 5, verbose = True, lda_model = False):
  '''
  Given two datasets are:
    1) merged with dataset_lda functions
    2) the merged dataset is tranformed according to LDA(1) values with the function lda_comp
    3) a binary classifier (KNN) with n_neigh (deafult = 5) id fitted and the accuracy of prediction is saved.
       if accuracy is weighted by the complement of the frequency of that class: f_i = 0.8 --> w_i = 1 - 0.8 = 0.2 
    4) accuracy test and C.I are compured trough the function accuracy_test

    --> accuracy and C.I. from accuracy_test are returned from this function

     - if lda_model = true, beside accuracy and C.I. are returned also the fitted lda and the dataframe compute with lda_comp
     - if verbose = True a summary plot is computed
   
  '''
  np.random.seed(123)
  lda, data = lda_comp(data1,data2)
  X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1], data.loc[:,'labels'])
  #try:
  knn = KNeighborsClassifier(n_neigh)
  knn.fit(X_train, y_train)
  res = knn.predict(X_test)
  data1_freq = round(y_train[y_train == 1.0].count()/len(y_train),2)
  sample_weight = [data1_freq, 1-data1_freq]
  sample_weihgt_array = np.array([sample_weight[int(el)] for el in y_test])
  acc = accuracy_score(y_test,res,sample_weight = sample_weihgt_array)
  _, conf = accuracy_test(acc, len(X_test), null_mean, verbose = verbose)

  if lda_model:
    return acc, conf, lda, data
  else:
    return acc, conf
    
  #except:
  #  acc = 0.0
  # _, conf = accuracy_test(acc, len(X_test), null_mean, verbose = verbose)
  #  if lda_model:
  #    return acc, conf, lda, data
  #  else:
  #    return acc, conf

def compute_respective_year(df, date_column):
  '''
  Given a Dataframe and its date column in datetime format, parse each date as MM-DD.
  es: 2018-09-02 --> 9-2
  '''
  
  m_d = df[date_column].apply(lambda x: str(x.month)+'-'+str(x.day)).to_numpy()
  return m_d

```

```python
def compute_masked_correlation_diff(data1,data2,columns, alpha = 0.4 , inspect = False, verbose = True):
  '''
  Given two dataset, selecet the columns. The difference of 
  the correlation marix respectively computed and is chechked if the correlation is variated more than alpha value.
  A df encoding problematic correlation is returned
  if inspect = True a more detailed df is returned, containing also the past and current correlation value
  if verbose = True a output is printed 
  '''
  corr1 = data1[columns].corr()
  corr2 = data2[columns].corr()
  corr = corr1 - corr2
  col_i, col_j = np.where(np.abs(corr) > alpha)
  map_col = dict(zip(np.arange(len(columns)+1),columns))
  df = []
  df.append([map_col[el] for el in col_i])
  df.append([map_col[el] for el in col_j])
  df.append(np.array([corr.iloc[col_i[el],col_j[el]] for el in range(len(col_i))]))
  df_columns = ['feature_1', 'feature_2', 'corr']

  if inspect:
    df.append(np.array([corr1.iloc[col_i[el],col_j[el]] for el in range(len(col_i))]))
    df.append(np.array([corr2.iloc[col_i[el],col_j[el]] for el in range(len(col_i))]))
    df_columns += ['past','current']

  df = pd.DataFrame(df).T
  df.columns = df_columns
  

  if verbose:
    plot_matrix(corr,col, title = 'CORRELATION DIFFERENCE')
    if len(df) > 0: #if there are problematic correlations
      for el in range(len(df)):
        print(f'''Detected significant change wrt {df.iloc[el,:]['feature_1']} vs' {df.iloc[el,:]['feature_2']}
        correlation difference = {df.iloc[el,:]['corr']} /n''')
        if inspect:
          print(f'''past correlation = {df.iloc[el,:]['past']}, current correlation = {df.iloc[el,:]['current']} /n''')
    else:
      print('No significant change has been detected')

  return df

```
